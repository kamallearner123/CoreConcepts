```python
count = 0
def permute(nums, digits, path=[]):
    global count
    #print(path)
    if not nums or len(path)==digits:
        count += 1
        print(f"Comd => {path}, count => {count}")
        yield path
    for i in range(len(nums)):
        yield from permute(nums[:i] + nums[i+1:], digits, path + [nums[i]])

nums = [1, 2, 3, 4, 5]

iter1 = permute(nums, 3)

for comb in iter1:
    print(comb)
```
