```python
import itertools
print(help(itertools.combinations))
print(help(itertools.permutations))
```

```python
gen1 = itertools.permutations(["chicken roti","biryani", "starter", "drink"], 2)
for comb in gen1:
    print(comb)
```

```python
gen2 = itertools.batched(["kamal","mounisha", "pavani", "ram", "david", "sri"], 2)
for comb in gen2:
    print(comb)
```
