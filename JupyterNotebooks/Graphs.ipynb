```python
# Graphs
def floyd_warshall(graph, num_vertices):
    dist = [[float('inf')] * num_vertices for _ in range(num_vertices)]
    
    for u in graph:
        for v, w in graph[u]:
            dist[u][v] = w
    for i in range(num_vertices):
        dist[i][i] = 0

    for k in range(num_vertices):
        for i in range(num_vertices):
            for j in range(num_vertices):
                dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])
    
    return dist

graph = {0: [(1, 3), (2, 6)], 1: [(2, 2), (3, 1)], 2: [(3, 4)], 3: []}
print(floyd_warshall(graph, 4))
```

```python
import heapq
def dijkstra(graph, start):
    heap = [(0, start)]
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    while heap:
        (cost, node) = heapq.heappop(heap)
        for neighbor, weight in graph[node]:
            new_cost = cost + weight
            if new_cost < distances[neighbor]:
                distances[neighbor] = new_cost
                heapq.heappush(heap, (new_cost, neighbor))

    return distances

graph = {0: [(1, 4), (2, 1)], 1: [(3, 1)], 2: [(1, 2), (3, 5)], 3: []}
print(dijkstra(graph, 0))
```

```python
# Weighted Directed Graph

import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add weighted edges (node1, node2, weight)
edges = [
    ('A', 'B', 3),
    ('A', 'C', 2),
    ('B', 'C', 5),
    ('C', 'D', 4),
    ('D', 'A', 1)
]

# Add edges to the graph
for u, v, w in edges:
    G.add_edge(u, v, weight=w)

# Get positions for nodes
pos = nx.spring_layout(G)

# Draw nodes and edges
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='red', node_size=2000, font_size=12, arrows=True)

# Draw edge labels (weights)
edge_labels = {(u, v): w for u, v, w in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

# Show plot
plt.show()

```
